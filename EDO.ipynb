# Problemas de Ecuaciones Diferenciales

## Retratos de Fase y Sistemas Asociados

### Instrucciones
1. Determinar un sistema de ecuaciones asociado para cada una de las ecuaciones diferenciales dadas.
2. Realizar el retrato de fase del sistema para cada caso.

Las ecuaciones diferenciales son:
- \(2x'' - x' + x = 0\)
- \(x'' + x' + x = 0\)
- \(x'' - 2x' + x = 0\)


# Ejercicio 1



$$
2x'' - x' + x = 0
$$


## Solución

### Introducción de variables de estado

Para convertir la ecuación diferencial de segundo orden en un sistema de primer orden, realizamos los siguientes pasos:

Definimos las variables de estado para reducir el orden de la ecuación diferencial:

$$
y_1 = x, \quad y_2 = x'
$$

### Derivación de las primeras derivadas

Expresamos las primeras derivadas de estas nuevas variables:

$$
\dot{y}_1 = y_2
$$

$$
\dot{y}_2 = x''
$$

### Sustitución en la ecuación original

Sustituimos $$ y_1 $$ y $$ \dot{y}_2 $$ en la ecuación original para encontrar una expresión para \(x''\):

$$
2x'' - x' + x = 0
$$

Reorganizamos términos para despejar \(x''\):

$$
2x'' = x' - x
$$

$$
x'' = \frac{1}{2} (x' - x)
$$

Sustituimos \(x'\) y \(x\) por \(y_2\) y \(y_1\) respectivamente:

$$
\dot{y}_2 = \frac{1}{2} (y_2 - y_1)
$$

### Formulación del sistema de ecuaciones

Finalmente, el sistema de ecuaciones asociado a la ecuación dada es:

$$
\begin{cases}
\dot{y}_1 = y_2 \\
\dot{y}_2 = \frac{1}{2} (y_2 - y_1)
\end{cases}
$$

Este sistema representa la forma de primer orden de la ecuación diferencial original.


## Plantamiento del Retrato de Fase


### Sistema de Ecuaciones de Primer Orden

El sistema de ecuaciones derivado de la ecuación diferencial original es:

$$
\begin{cases}
\dot{y}_1 = y_2 \\
\dot{y}_2 = \frac{1}{2} (y_2 - y_1)
\end{cases}
$$

### Estabilidad y Comportamiento Dinámico

Al analizar el sistema en el espacio de fase, se pueden identificar puntos críticos o de equilibrio, donde las derivadas de todas las variables de estado se hacen cero. Para el sistema dado, el punto crítico es \( (0, 0) \). El análisis de estabilidad de este punto y las trayectorias en su entorno pueden revelar si el sistema tiende a estabilizarse, a diverger o a mantener oscilaciones periódicas.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def sistema(t, y):
    y1, y2 = y
    dydt = [y2, 0.5 * (y2 - y1)]
    return dydt

y1_values = np.linspace(-2, 2, 20)
y2_values = np.linspace(-2, 2, 20)


plt.figure(figsize=(8, 6))

for y1 in y1_values:
    for y2 in y2_values:
        sol = solve_ivp(sistema, [0, 10], [y1, y2], t_eval=np.linspace(0, 10, 100))
        plt.plot(sol.y[0], sol.y[1], color='blue')

plt.title('Retrato de fase para $2x\'\' - x\' + x = 0$')
plt.xlabel('$y_1$ (posición)')
plt.ylabel('$y_2$ (velocidad)')
plt.grid(True)

plt.show()
```

## Análisis del Comportamiento del Sistema

El sistema de ecuaciones derivado de la ecuación diferencial \(2x'' - x' + x = 0\) se representa en forma matricial como:

$$
\begin{pmatrix}
\dot{y}_1 \\
\dot{y}_2
\end{pmatrix} =
\begin{pmatrix}
0 & 1 \\
-0.5 & 0.5
\end{pmatrix}
\begin{pmatrix}
y_1 \\
y_2
\end{pmatrix}
$$

### Cálculo de Eigenvalores

Resolviendo la ecuación característica $$ 0.5\lambda^2 - 0.5\lambda + 0.5 = 0 $$ , obtenemos dos eigenvalores complejos conjugados con partes reales positivas, lo que indica un foco.

### Interpretación de los Eigenvalores

Los eigenvalores complejos sugieren que las soluciones del sistema exhibirán un comportamiento oscilatorio amortiguado hacia el punto de equilibrio \( (0,0) \).

### Conclusión

Este comportamiento oscilatorio y convergente se visualiza en el retrato de fase, mostrando cómo todas las trayectorias espirales hacia el origen, reflejando la naturaleza amortiguada del sistema.

### Sustitución en la ecuación original

$$
x'' + x' + x = 0

## Plantamiento del Retrato de Fase



### Sistema de Ecuaciones de Primer Orden

El sistema de ecuaciones derivado de la ecuación diferencial original es:

$$
\begin{cases}
\dot{y}_1 = y_2 \\
\dot{y}_2 = -y_2 - y_1
\end{cases}
$$

### Estabilidad y Comportamiento Dinámico

Es crucial identificar los puntos de equilibrio, donde las derivadas de todas las variables de estado se hacen cero. Para este sistema, el único punto crítico es \( (0, 0) \). El análisis de estabilidad de este punto y las trayectorias en su entorno determinan si el sistema tiende hacia la estabilidad, diverge o mantiene oscilaciones.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def sistema(t, y):
    y1, y2 = y
    dydt = [y2, -y1 - y2]
    return dydt

y1_values = np.linspace(-2, 2, 20)
y2_values = np.linspace(-2, 2, 20)

plt.figure(figsize=(8, 6))

for y1 in y1_values:
    for y2 in y2_values:
        sol = solve_ivp(sistema, [0, 10], [y1, y2], t_eval=np.linspace(0, 10, 100))
        plt.plot(sol.y[0], sol.y[1], color='green')

plt.title('Retrato de fase para $x\'\' + x\' + x = 0$')
plt.xlabel('$y_1$ (posición)')
plt.ylabel('$y_2$ (velocidad)')
plt.grid(True)

plt.show()
```

## Análisis del Comportamiento del Sistema para \(x'' + x' + x = 0\)

El sistema de ecuaciones derivado se representa en forma matricial como:

$$
\begin{pmatrix}
\dot{y}_1 \\
\dot{y}_2
\end{pmatrix} =
\begin{pmatrix}
0 & 1 \\
-1 & -1
\end{pmatrix}
\begin{pmatrix}
y_1 \\
y_2
\end{pmatrix}
$$

### Cálculo de Eigenvalores

Resolviendo la ecuación característica $$ \lambda^2 + \lambda + 1 = 0 $$, obtenemos eigenvalores complejos conjugados con partes reales negativas, indicando un espiral estable.

### Interpretación de los Eigenvalores

Los eigenvalores complejos con parte real negativa sugieren que las soluciones del sistema oscilarán hacia el equilibrio en espiral, indicando un comportamiento oscilatorio amortiguado.

### Conclusión

El retrato de fase muestra trayectorias que espiralan hacia el punto de equilibrio \( (0,0) \), lo que refleja la dinámica estable y oscilatoria del sistema bajo la influencia del amortiguamiento.

x'' - 2x' + x = 0

## Plantamiento Retrato de Fase


### Sistema de Ecuaciones de Primer Orden

El sistema de ecuaciones derivado de la ecuación diferencial original es:

$$
\begin{cases}
\dot{y}_1 = y_2 \\
\dot{y}_2 = 2y_2 - y_1
\end{cases}
$$

### Estabilidad y Comportamiento Dinámico

Se investiga el espacio de fase para identificar puntos de equilibrio, donde las derivadas de las variables de estado se anulan. Para este sistema, el único punto crítico es \( (0, 0) \). El sistema tiende hacia este punto de equilibrio de manera más rápida, mostrando así un comportamiento oscilatorio.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def sistema(t, y):
    y1, y2 = y
    dydt = [y2, 2*y2 - y1]
    return dydt

y1_values = np.linspace(-2, 2, 20)
y2_values = np.linspace(-2, 2, 20)


plt.figure(figsize=(8, 6))

for y1 in y1_values:
    for y2 in y2_values:
        sol = solve_ivp(sistema, [0, 10], [y1, y2], t_eval=np.linspace(0, 10, 100))
        plt.plot(sol.y[0], sol.y[1], color='red')

plt.title('Retrato de fase para $x\'\' - 2x\' + x = 0$')
plt.xlabel('$y_1$ (posición)')
plt.ylabel('$y_2$ (velocidad)')
plt.grid(True)

plt.show()
```

## Análisis del Comportamiento del Sistema

El sistema de ecuaciones derivado de la ecuación diferencial \(x'' - 2x' + x = 0\) se representa en forma matricial como:

$$
\begin{pmatrix}
\dot{y}_1 \\
\dot{y}_2
\end{pmatrix} =
\begin{pmatrix}
0 & 1 \\
-1 & 2
\end{pmatrix}
\begin{pmatrix}
y_1 \\
y_2
\end{pmatrix}
$$

### Cálculo de Eigenvalores

Resolviendo la ecuación característica $$\ \lambda^2 - 2\lambda + 1 = 0 \$$, obtenemos un eigenvalor repetido $ \( \lambda = 1 \) $, indicando un nodo estrella.

### Interpretación de los Eigenvalores

Los eigenvalores reales y positivos indican que el sistema es críticamente amortiguado. Las trayectorias se aproximan al punto de equilibrio \( (0, 0) \) de manera rápida, lo que refleja la naturaleza dinámica estable del sistema bajo condiciones específicas.

### Conclusión

El retrato de fase muestra trayectorias que convergen hacia el punto de equilibrio, demostrando la complejidad dinámica de un sistema que, a primera vista, podría parecer simple.
